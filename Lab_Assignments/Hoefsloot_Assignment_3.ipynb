```python
#Function that takes as input 1 image x and the cluster centroids mu
#It returns the index of the closest centroid
def compute_centroid(x,mu):
    xr = np.reshape(x,(64,1))
    y = np.subtract(mu, xr)  #The difference between the datapoints and the mean
    z = numpy.linalg.norm(y, ord=None, axis=0, keepdims=False)  #Take the norm over the columns
    return numpy.argmin(z)   #Return the index of the closest cluster

##TEST VALUES
#mu = np.random.uniform(0,10,(64,10))
#x = np.reshape(digits.images[:1],(1,64)) 
#compute_centroid(x,mu)
```

```python
#Function that takes as input;: a cluster number, the cluster centroids
#and all the datapoints assigned to that cluster
#It returns an updated mu matrix for that cluster number
def update_centroid(centroid, mu, X_c):
    aver = numpy.mean(X_c, axis=0) #Take the average over the columns in X_c
    mu[:, centroid] = aver  #Change only 1 specific column in mu
    return mu

##TEST VALUES
#mu = np.random.uniform(0,16,(64,10))
#x = np.reshape(digits.images[:1500],(1500,64)) 
#c = np.random.randint(0,k,(1500))
#test_X_c = x[c==3]
#update_centroid(3,mu,test_X_c)
```

```python
#The cost function
#Returns the minimum of the centroid
def cost_function(x,mu):
    xr = np.reshape(x,(64,1))
    y = np.subtract(mu, xr)   #The difference between the datapoints and the mean
    z = numpy.linalg.norm(y, ord=None, axis=0, keepdims=False)  #Take the norm over the columns
    return z[numpy.argmin(z)] #The minimum of the centroid

##TEST VALUES
#mu = np.random.uniform(0,16,(64,10))
#x = np.reshape(digits.images[:1],(1,64)) 
#cost_function(x,mu)
```

```python
#The amount of iterations
initialisation_iterations = 10
iterations = 100

k = 10  #The amount of clusters k
best_cost = 100  #Initializations of the best cost

x = np.reshape(digits.images[:1500],(1500,64))
x_test = np.reshape(digits.images[1500:],(297,64))


iters = x.shape[0]
x_axis = range(iters)
for h in range(initialisation_iterations):  #Different initialisations of mu
    mu = np.random.uniform(0,16,(64,k))  #Centroid matrix   
    c = np.random.randint(0,k,(1500))    #The vector that assigns each digit to a centroid
    cost_plot = []    #Initialization of the list I'll use to plot the cost
    for j in range(iterations):    #Amount of iterations of the k means algorithm
        co = 0    #The current cost
        for i in range(k):   #Updates the centroid of all the clusters
            mu = update_centroid(i, mu, x[c==i])
        for i in range(x.shape[0]):   #Assigns a new cluster to each datapoint
            c[i] = compute_centroid(x[i,:],mu)
            co += cost_function(x[i,:],mu)
        co = co/iters   #The cost after the updates
        cost_plot.append(co)  #Append the cost to the list
        if co < best_cost:    #Keeps track of the best cost, mu and c
            best_cost = co
            best_mu = mu
            best_c = c
    plt.plot(range(iterations),cost_plot)  #Plot the full list of the costs per initialisations of mu
    print "In initialization iteration ", h+1, ", the costfunction is:"
    plt.show()


print "The visualizations of the best result:"
visualize(best_c)
visualize_centroids(best_mu)
```
